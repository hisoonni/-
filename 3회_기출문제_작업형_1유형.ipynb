# **[작업형 1-1]**
2022년 데이터 중 2022년 중앙값보다 큰 값의 데이터 수¶
- data: t1-data2.csv

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/빅데이터분석기사/t1-data2.csv", index_col='year')
df
```

```python
# solution
# 중앙값 변수 만들기
data = df.loc["2022년"].median()
```

```python
# 중앙값 보다 큰 값의 데이터 수를 출력하기
print(sum(df.loc["2022년",:] > data))
```

결측치 데이터(행)을 제거하고, 앞에서부터 60% 데이터만 활용해, 'f1' 컬럼 3사분위 값을 구하시오
- 60%가 소수점일 경우 절사(예: 36.6 일때 36으로 계산)
- data: t1-data1.csv

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/빅데이터분석기사/t1-data1.csv")
df
```

```python
import pandas as pd

# CSV 파일에서 데이터 프레임으로 데이터 읽기
df = pd.read_csv("/content/drive/MyDrive/빅데이터분석기사/t1-data1.csv")

# 결측치가 있는 행 제거
df = df.dropna()

# 데이터 프레임의 앞에서부터 60%의 데이터만 선택 (소수점인 경우 절사)
n = int(len(df) * 0.6)
df = df.head(n)

# 'f1' 컬럼의 3사분위 값 계산
q3 = df['f1'].quantile(0.75)

# 3사분위 값을 출력
print(q3)
```

```python
#결측치 제거
df = df.dropna()

#60%가 소수점일 경우 절사(예: 36.6 일때 36으로 계산)
df = df.iloc[:int(len(df)*0.6)]

#결과값 출력
print(df['f1'].quantile(.75))
```
