## 결측치 처리
-  주어진 데이터에서 결측치가 80%이상 되는 컬럼은(변수는) 삭제하고, 80% 미만인 결측치가 있는 컬럼은 'city'별 중앙값으로 값을 대체하고 'f1'컬럼의 평균값을 출력하세요!
- 데이터셋 : basic1.csv 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작
- 강의 영상 : https://youtu.be/WqlpqBRn7x4

```python
# 라이브러리 및 데이터 불러오기

import pandas as pd
import numpy as np
df=pd.read_csv("/kaggle/input/bigdatacertificationkr/basic1.csv")
df
```

```python
# EDA - 결측값 확인(비율 확인)

# 결측치 비율 확인
missing_ratio = df.isnull().sum() / df.shape[0]
missing_ratio
```

```python
# 80%이상 결측치 컬럼, 삭제
# 결측치가 80% 이상인 컬럼 삭제
columns_to_drop = missing_ratio[missing_ratio >= 0.8].index
df = df.drop(columns_to_drop, axis=1)
```

```python
# 80%미만 결측치 컬럼, city별 중앙값으로 대체
for column in df.columns:
    if column != 'f1' and column != 'city':
        if missing_ratio[column] < 0.8 and df[column].isnull().sum() > 0:
            median_map = df.groupby('city')[column].transform(lambda x: np.median(x))
            df[column] = np.where(df[column].isnull(), median_map, df[column])
```

```python
# f1 평균값 결과 출력

f1_mean = np.mean(df['f1'])
print(f1_mean)
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
df.head()
```

```python
# EDA - 결측비율 확인
df.isnull().sum()/df.shape[0]
```

```python
# f3 컬럼 삭제
print("삭제 전:", df.shape)
df = df.drop(['f3'], axis=1)
print("삭제 후:", df.shape)

# 결측치가 있는 컬럼을 제거하는 2가지 방법
# df.drop(['B', 'C'], axis=1)
# df.drop(columns=['B', 'C'])
```

```python
# 도시 확인
df['city'].unique()
```

```python
# 도시별 중앙값 계산
s=df[df['city']=='서울']['f1'].median()
k=df[df['city']=='경기']['f1'].median()
b=df[df['city']=='부산']['f1'].median()
d=df[df['city']=='대구']['f1'].median()
s, k, b, d

#방법2
# k, d, b, s = df.groupby('city')['f1'].median()
```

```python
# 대체 전 데이터 샘플 출력
df[18:21]
```

```python
# f1결측치 city별 중앙값으로 대체
df['f1'] = df['f1'].fillna(df['city'].map({'서울':s,'경기':k,'부산':b,'대구':d}))

# 만약 그냥 f1 중앙값으로 대체 한다면 
# df['f1'] = df['f1'].fillna(df['f1'].median())
```
