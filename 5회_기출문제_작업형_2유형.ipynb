```python
# 예측할 타켓 price 을 train 데이터로 옮기기
y_train = train.pop("price")
```

# **수치형 활용**

```python
cols = [ 'year', 'mileage', 'tax', 'mpg', 'engineSize']
train = train[cols]
test = test[cols]
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, y_train, test_size=0.2, random_state=2022)
X_tr.shape, X_val.shape, y_tr.shape, y_val.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error

def rmse(y_true, y_pred) :
  return mean_squared_error(y_true, y_pred) ** 0.5
  # return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
rmse(y_val, pred)
```

# **수치형 + 범주형 활용**

```python
train = pd.read_csv("/content/drive/MyDrive/빅데이터분석기사/5회/train.csv")
test =pd.read_csv("/content/drive/MyDrive/빅데이터분석기사/5회/test.csv")
```

```python
# 원 핫 인코딩
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
train.shape, test.shape
```

```python
train.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, y_train, test_size=0.2, random_state=2022)
X_tr.shape, X_val.shape, y_tr.shape, y_val.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
rmse(y_val, pred)
```

# **Test 예측**

```python
pred = rf.predict(test)
result = pd.DtaFrame({
    'pred':pred
})
result.to_csv("result.csv", index=False)
```
