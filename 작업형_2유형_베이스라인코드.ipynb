```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('/content/sample_data/train.csv')
test=pd.read_csv('/content/sample_data/test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
train=train.drop('ID',axis=1)

test=test.drop('ID',axis=1)

train_y=train['Segmentation']

#수치형 컬럼 분류 -> 스케일링 적용
col_num=['Age','Work_Experience','Family_Size']
#범주형 컬럼 분류 -> 원핫인코딩 적용
col_cat=['Gender','Ever_Married','Graduated','Profession','Spending_Score','Var_1']
```

```python
train.head()
```

```python
from pandas import get_dummies
#범주형 자료는 원핫인코딩
train_encoding=pd.get_dummies(train[col_cat])

train_encoding.head()
```

```python
test_encoding=pd.get_dummies(test[col_cat])

test_encoding.head()
```

```python
train[col_num].head()
```

```python
#수치형 자료 스케일링 적용
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

scaler.fit(train[col_num])

train[col_num]=scaler.transform(train[col_num])

test[col_num]=scaler.transform(test[col_num])

train=pd.concat([train[col_num],train_encoding],axis=1)
```

```python
train.head()
```

```python
test=pd.concat([test[col_num],test_encoding],axis=1)

test.head()
```

```python
# 데이터셋 분리
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train, train_y, test_size=0.3, random_state=42)

X_tr.head()

X_val.head()

y_tr.head()

y_val.head()
```

```python
# 모델 생성 ( 랜더포레스트로 무조건 )
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()

rfc.fit(X_tr, y_tr)

predict_y_tr=rfc.predict(X_tr)

predict_y_tr
```

```python
from sklearn.metrics import f1_score

f1_score(y_tr, predict_y_tr, average='macro')
```

```python
predict_y_val=rfc.predict(X_val)

predict_y_val
```

```python
f1_score(y_val, predict_y_val, average='macro')

predict_y_test=rfc.predict(test)

predict_y_test
```

```python
predict_testpd=pd.DataFrame(predict_y_test)
```

```python
predict_testpd.to_csv('수험번호.csv', index=False)
```
